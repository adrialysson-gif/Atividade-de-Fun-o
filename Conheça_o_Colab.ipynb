```
def livro(nome, qtdPaginas, autor, preco):

    print("\n Inf do Livro ")
    print(f"Nome: {nome}")
    print(f"Quantidade de páginas: {qtdPaginas}")
    print(f"Autor: {autor}")
    print(f"Preço: R$ {preco:.2f}")


def menu():
    print("=== Cadastro de Livros ===")
    nome = input("Digite o nome do livro que voce deseja procurar: ")
    qtdPaginas = int(input("Digite a quantidade de páginas que o livro possue: "))
    autor = input("Digite o nome do autor do livro : ")
    preco = float(input("Digite o preço do livro: "))

    livro(nome, qtdPaginas, autor, preco)


menu()
```

```
def alterarNome(nome):
    novo_nome = input("Digite o novo nome do aluno: ")
    return novo_nome

def alterarMatricula(matricula):
    nova_matricula = input("Digite a matrícula atualizada do aluno: ")
    return nova_matricula

# Programa principal
nome = input("Digite o nome do aluno: ")
matricula = input("Digite a matrícula do aluno: ")

print("\nDados cadastrados:")
print(f"Nome: {nome}")
print(f"Matrícula: {matricula}")

while True:
    alterar = input("\nDeseja alterar algum dado? (s/n): ").lower()

    if alterar == 's':
        opcao = input("Digite o que deseja alterar (nome/matricula): ").lower()

        if opcao == "nome":
            nome = alterarNome(nome)
        elif opcao == "matricula":
            matricula = alterarMatricula(matricula)
        else:
            print("Opção inválida.")
            continue

        print("\nDado alterado com sucesso!")
        print(f"Nome: {nome}")
        print(f"Matrícula: {matricula}")
    elif alterar == 'n':
        print("\nEncerrando o programa.")
        break
    else:
        print("Digite apenas 's' para sim ou 'n' para não.")
```

```
def soma(a, b, c):
    """Função que retorna a soma de três números"""
    return a + b + c

def media(func_soma, a, b, c):
    """Função que calcula a média usando a função soma"""
    return func_soma(a, b, c) / 3

def menu():
    """Função principal que recebe os números do usuário"""
    while True:
        try:
            print("\n--- Cálculo de Soma e Média ---")
            a = float(input("Digite o primeiro número: "))
            b = float(input("Digite o segundo número: "))
            c = float(input("Digite o terceiro número: "))

            resultado_soma = soma(a, b, c)
            resultado_media = media(soma, a, b, c)

            print(f"\nA soma dos números é: {resultado_soma}")
            print(f"A média dos números é: {resultado_media:.2f}")

            continuar = input("\nDeseja calcular novamente? (s/n): ").lower()
            if continuar != 's':
                print("finalizando o programa...")
                break

        except ValueError:
            print("Erro: digite apenas números válidos!")


menu()
```
